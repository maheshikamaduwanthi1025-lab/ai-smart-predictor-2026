```python
%%writefile myuniqueapp.py
import streamlit as st
import pandas as pd
import numpy as np

import gdown
import zipfile
import os

# 1. Google Drive File ID එක සහ Download කරගන්නා ZIP file එකේ නම
file_id = "1rTj8g3_fB0f-KT5q8mWP5Ol99VMea6K4"
url = f"https://drive.google.com/uc?id={file_id}"
zip_output = "downloaded_dataset.zip"

print("Downloading file from Google Drive...")

# 2. File එක Google Drive එකෙන් download කිරීම
gdown.download(url, zip_output, quiet=False)

# 3. Extract කරන්න අවශ්‍ය ෆෝල්ඩරයේ නම
extract_folder = "dataset"

print("\nExtracting ZIP file...")
try:
    # Zip file එක extract කිරීම
    with zipfile.ZipFile(zip_output, "r") as zip_ref:
        zip_ref.extractall(extract_folder)
    print(f"සාර්ථකව extract කරලා ඉවරයි! Files සේව් වුනේ '{extract_folder}' ෆෝල්ඩරයට.\n")

    # 4. Extract වුණු ෆෝල්ඩරය ඇතුළේ තියෙන files ලැයිස්තුව පෙන්වීම
    print("Files inside dataset folder:\n")
    for root, dirs, files in os.walk(extract_folder):
        for file in files:
            print(os.path.join(root, file))

except zipfile.BadZipFile:
    print("Error: Download වුණු file එක valid ZIP file එකක් නෙවෙයි. (Google Drive file එක public share කරලා තියෙනවාද කියා පරීක්ෂා කරන්න).")
```
